```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
df = pd.read_csv('/Users/purva/Desktop/Data Science Boot Camp/examples/avocado 2.csv')
```

## Plotting one variable in 4 different ways

```python
plt.hist('AveragePrice')
plt.show()

## one variable - Graph 1
```

```python
plt.plot(df['AveragePrice'])
plt.show()
## One variable - Graph 2
```

```python
sns.distplot(df['AveragePrice'])
plt.show()
```

```python
sns.distplot(df['AveragePrice'], hist=False, rug=True, color="r")
```

## Plotting two continuous variable in 3 different ways


```python
plt.scatter(df['Total Volume'],df['AveragePrice'])
plt.show()
## Two continuous variable - Graph 1
```

```python
sns.lineplot(df['Total Volume'],df['AveragePrice'], data=df)
plt.show()
```

```python
sns.kdeplot(df['Total Volume'],df['AveragePrice'])
```

## Plotting one categorical and one continuous variable in 6 different ways

```python
sns.boxplot(df['type'], df['AveragePrice'], data=df)
plt.show()
```

```python
sns.violinplot(df['type'], df['AveragePrice'],data=df)
plt.show()
```

```python
sns.barplot(df['type'], df['AveragePrice'], data=df)
plt.show()
```

```python
sns.pointplot(df['type'], df['AveragePrice'], data=df)
plt.show()
```

```python
sns.stripplot(df['type'], df['AveragePrice'], data=df)
plt.show()
```

```python
sns.boxenplot(df['type'], df['AveragePrice'], data=df)
plt.show()
```
